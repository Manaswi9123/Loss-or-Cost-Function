```python
import numpy as np
yp=np.array([1,1,0,0,1])
yt=np.array([0.30,0.7,1,0,0.5])
```

```python
def mae(yt,yp):
    total_err=0
    for tr,pr in zip(yt,yp):
        total_err+=abs(tr-pr)
    print('Total Error:',total_err)
    me=total_err/len(yt)
    print('MAE:',me)
    return me
```

```python
mae(yt,yp)
```

```python
np.mean(np.abs(yp-yt))
```

```python
ypn=[min(i,1-e) for i in ypn]
ypn
```

```python
ypn=np.array(ypn)
np.log(ypn)
```

```python
-np.mean(yt*np.log(ypn)+(1-yt)*np.log(1-ypn))
```

```python
def logLoss(yt,yp):
    ypn=[max(i,e) for i in yp]
    ypn=[min(i,1-e) for i in ypn]
    ypn=np.array(ypn)
    return -np.mean(yt*np.log(ypn)+(1-yt)*np.log(1-ypn))
```

```python
logLoss(yt,yp)
```

```python
def mse(yt,yp):
    total_err=0
    for tr,pr in zip(yt,yp):
        ms=(tr-pr)**2
        total_err+=ms
    print('Total Error:',total_err)
    me=total_err/len(yt)
    print('MSE:',me)
    return me
```

```python
mse(yt,yp)
```
